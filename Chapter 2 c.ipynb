```python
from sklearn.datasets import make_blobs
X, y = make_blobs(centers=4, random_state = 8)
y_c = y % 2
fig, ax = plt.subplots(1,2,figsize=(16,6))
ax[0].scatter(X[:,0], X[:,1], c = y)
ax[1].scatter(X[:,0], X[:,1], c = y_c)
plt.show()
print(X[:5])
print(y[:5])
print(y_c[:5])
```

```python
lsvc = svm.LinearSVC().fit(X, y_c)

from mglearn.plots import plot_2d_separator
plot_2d_separator(lsvc, X)
plt.scatter(X[:,0], X[:,1], c = y_c)
```

```python
from mpl_toolkits.mplot3d import Axes3D

X_ext = np.concatenate((np.transpose([X[:,0]]),np.transpose([X[:,1]]), np.transpose([X[:,1] ** 2])), axis = 1)
X_ext = np.hstack((X, X[:,1:]**2))

fig = plt.figure(figsize=(12,12))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X_ext[:,0],X_ext[:,1],X_ext[:,2],c=y_c)
```

```python
linear_svm_3d = svm.LinearSVC().fit(X_ext, y_c)
coef, intercept = linear_svm_3d.coef_.ravel(), linear_svm_3d.intercept_

# show linear decision boundary
figure = plt.figure()
ax = Axes3D(figure, elev=-152, azim=-26)
xx = np.linspace(X_ext[:, 0].min() - 2, X_ext[:, 0].max() + 2, 50)
yy = np.linspace(X_ext[:, 1].min() - 2, X_ext[:, 1].max() + 2, 50)
mask = y_c == 0

XX, YY = np.meshgrid(xx, yy)
ZZ = (coef[0] * XX + coef[1] * YY + intercept) / -coef[2]
ax.plot_surface(XX, YY, ZZ, rstride=8, cstride=8, alpha=0.3)
ax.scatter(X_ext[mask, 0], X_ext[mask, 1], X_ext[mask, 2], c='b',
           cmap=mglearn.cm2, s=60)
ax.scatter(X_ext[~mask, 0], X_ext[~mask, 1], X_ext[~mask, 2], c='r', marker='^',
           cmap=mglearn.cm2, s=60)
```

```python
print(svm.LinearSVC().fit(X, y_c).score(X, y_c))
print(svm.LinearSVC().fit(X_ext, y_c).score(X_ext, y_c))
```

```python
ZZ = YY ** 2
dec = linear_svm_3d.decision_function(np.c_[XX.ravel(), YY.ravel(), ZZ.ravel()])
plt.contourf(XX, YY, dec.reshape(XX.shape), levels=[dec.min(), 0, dec.max()],
             cmap=mglearn.cm2, alpha=0.5)
mglearn.discrete_scatter(X[:, 0], X[:, 1], y_c)
plt.xlabel("Feature 0")
plt.ylabel("Feature 1")
```

```python
from sklearn.svm import SVC
X, y = mglearn.tools.make_handcrafted_dataset()
svm = SVC(kernel='rbf', C=10, gamma=0.1).fit(X, y)
mglearn.plots.plot_2d_separator(svm, X, eps=.5)
mglearn.discrete_scatter(X[:, 0], X[:, 1], y)
# plot support vectors
sv = svm.support_vectors_
# class labels of support vectors are given by the sign of the dual coefficients
sv_labels = svm.dual_coef_.ravel() > 0
mglearn.discrete_scatter(sv[:, 0], sv[:, 1], sv_labels, s=15, markeredgewidth=3)
plt.xlabel("Feature 0")
plt.ylabel("Feature 1")
```

```python
def scale_data2(*args):
    mean = np.vstack(args).mean(axis=0)
    std = np.vstack(args).std(axis=0)
    results = []
    for arg in args:
        results.append((arg - mean) / std)
    return results

X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target, random_state = 0)
X_tr_sc2, X_te_sc2 = scale_data2(X_train, X_test)
```
