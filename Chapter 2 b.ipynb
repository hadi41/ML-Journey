```python
import mglearn
from sklearn.model_selection import train_test_split
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()
display(pd.DataFrame(np.concatenate((np.transpose([cancer.target]),cancer.data, ),axis=1), columns = np.concatenate((['malignant or benign'],cancer.feature_names))))
from sklearn.datasets import load_boston
boston = load_boston()
display(pd.DataFrame(np.concatenate((np.transpose(np.array([boston.target])),boston.data),axis=1), columns=np.concatenate((['price'],boston.feature_names))))
```

```python
from sklearn.tree import DecisionTreeClassifier

X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target, random_state = 0)
```

```python
tree = DecisionTreeClassifier()
tree.fit(X_train, y_train)
print('Train:\t',tree.score(X_train, y_train))
print('Test:\t',tree.score(X_test, y_test))
```

```python
tree = DecisionTreeClassifier(max_depth = 1)
tree.fit(X_train, y_train)
print('Train:\t',tree.score(X_train, y_train))
print('Test:\t',tree.score(X_test, y_test))
```

```python
depths = np.arange(1,50)
train_scores = []
test_scores = []

for depth in depths:
    tree = DecisionTreeClassifier(max_depth = depth)
    tree.fit(X_train, y_train)
    train_scores.append(tree.score(X_train, y_train))
    test_scores.append(tree.score(X_test, y_test))
    
fig, ax = plt.subplots(1,1,figsize=(15,6))
ax.plot(depths, train_scores, label='Train')
ax.plot(depths, test_scores, label='Test')
ax.legend()
ax.set_xlabel('Max depth')
ax.set_ylabel('Accuracy')
ax.grid()
plt.show()

depths = np.arange(1,10)
train_scores = []
test_scores = []

for depth in depths:
    tree = DecisionTreeClassifier(max_depth = depth)
    tree.fit(X_train, y_train)
    train_scores.append(tree.score(X_train, y_train))
    test_scores.append(tree.score(X_test, y_test))
    
fig, ax = plt.subplots(1,1,figsize=(15,6))
ax.plot(depths, train_scores, label='Train')
ax.plot(depths, test_scores, label='Test')
ax.legend()
ax.set_xlabel('Max depth')
ax.set_ylabel('Accuracy')
ax.grid()
plt.show()
```

```python
def rank_feature_importance(feature_importances, feature_names):
    zipped = list(zip(feature_importances, feature_names))
    a = np.array(zipped)
    b = np.ma.masked_array(a, mask = (np.column_stack((a[:,0],a[:,0])) == '0.0'))
    c = b.compressed()
    d = []
    for i in range(round(len(c) / 2)):
        d.append([c[2*i], c[2*i+1]])
    d.sort()
    d.reverse()
    return d
```

```python
rank_feature_importance(tree.feature_importances_, cancer.feature_names)
```

```python
def plot_feature_importances(feature_importances, feature_names):
    labels = feature_names
    plt.barh(feature_names, feature_importances, height=0.8)
    plt.xlabel('Importance')
    plt.ylabel('Feature')
```

```python
fig, ax = plt.subplots(1,1,figsize=(15,7))
plot_feature_importances(tree.feature_importances_, cancer.feature_names)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import make_moons

X, y = make_moons(n_samples=100, noise=0.25, random_state=3)
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y,
                                                    random_state=42)

forest = RandomForestClassifier(n_estimators=5, random_state=2)
forest.fit(X_train, y_train)
```

```python
fig, axes = plt.subplots(2, 3, figsize=(20, 10))
for i, (ax, tree) in enumerate(zip(axes.ravel(), forest.estimators_)):
    ax.set_title("Tree {}".format(i))
    mglearn.plots.plot_tree_partition(X_train, y_train, tree, ax=ax)

mglearn.plots.plot_2d_separator(forest, X_train, fill=True, ax=axes[-1, -1],
                                alpha=.4)
axes[-1, -1].set_title("Random Forest")
mglearn.discrete_scatter(X_train[:, 0], X_train[:, 1], y_train)
```
