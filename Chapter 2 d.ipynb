```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import mglearn
```

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.datasets import make_blobs, make_circles
# X, y = make_blobs(centers = 3, n_samples = 100, random_state = 0, n_features = 2)
X, y = make_circles(factor = 0.5, n_samples = 100, noise = 0.25, random_state = 1)
print(y)
print([y])
y_named = np.array(['blue', 'red'])[y]
print(y_named)
```

```python
plt.scatter(X[:,0],X[:,1], c = y, cmap = 'bwr')
```

```python
X_train, X_test, y_train, y_test, y_train_named, y_test_named = train_test_split(X, y, y_named, random_state = 0)

gbc = GradientBoostingClassifier().fit(X_train, y_train_named)
print('Classes:\t',gbc.classes_)
```

```python
print('Train:\t', gbc.score(X_train, y_train_named))
print('Test:\t', gbc.score(X_test, y_test_named))
```

```python
# print('Decision function:\n',gbc.decision_function(X_test))
# print('Predictions:\n',gbc.predict(X_test))
print('Decision function vs prediction:\n',np.transpose(np.vstack((gbc.decision_function(X_test), gbc.predict(X_test)))))
```

```python
decisions = gbc.decision_function(X_test)
predictions = gbc.predict(X_test)
actual = y_test_named
data = {'Decision function': decisions, 'Prediction': predictions, 'Actual': actual}
table = pd.DataFrame(data = data)
display(table)
display(table.describe())
```

```python
fig, axes = plt.subplots(1, 2, figsize=(13, 5))
mglearn.tools.plot_2d_separator(gbc, X, ax=axes[0], alpha=.4,
                                fill=True, cm=mglearn.cm2)
scores_image = mglearn.tools.plot_2d_scores(gbc, X, ax=axes[1],
                                            alpha=.4, cm=mglearn.ReBl)

for ax in axes:
    # plot training and test points
    mglearn.discrete_scatter(X_test[:, 0], X_test[:, 1], y_test,
                             markers='^', ax=ax)
    mglearn.discrete_scatter(X_train[:, 0], X_train[:, 1], y_train,
                             markers='o', ax=ax)
    ax.set_xlabel("Feature 0")
    ax.set_ylabel("Feature 1")
cbar = plt.colorbar(scores_image, ax=axes.tolist())
axes[0].legend(["Test class 0", "Test class 1", "Train class 0",
                "Train class 1"], ncol=4, loc=(.1, 1.1))
```

```python
predictions = gbc.predict(X_test)
p = gbc.predict_proba(X_test)
confidence = [p[i,0] if predictions[i] == 'blue' else p[i,1] for i in range(p.shape[0])]
actual = y_test_named
data = {'Confidence': confidence, 'Predictions': predictions, 'Actual': actual}
table = pd.DataFrame(data = data)
display(table)
display(table.describe())
```

```python
from sklearn.preprocessing import StandardScaler
from matplotlib.colors import ListedColormap

ds = make_circles(factor = 0.5, n_samples = 100, noise = 0.25, random_state = 1)

X, y = ds
y_named = np.array(['red', 'blue'])[y]

X = StandardScaler().fit_transform(X)
X_train, X_test, y_train, y_test, y_train_named, y_test_named = \
    train_test_split(X, y, y_named, test_size=.4, random_state=42)

x_min, x_max = X[:, 0].min() - .5, X[:, 0].max() + .5
y_min, y_max = X[:, 1].min() - .5, X[:, 1].max() + .5

h = 0.02

xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))

gbc.fit(X_train, y_train_named)
score = gbc.score(X_test, y_test_named)

# Plot the decision boundary. For that, we will assign a color to each
# point in the mesh [x_min, x_max]x[y_min, y_max].

if hasattr(gbc, "decision_function"):
    Z = gbc.decision_function(np.c_[xx.ravel(), yy.ravel()])
else:
    Z = gbc.predict_proba(np.c_[xx.ravel(), yy.ravel()])[:, 1]

# Put the result into a color plot

cm = plt.cm.RdBu

Z = Z.reshape(xx.shape)
plt.contourf(xx, yy, Z, cmap='bwr', alpha=.8)

cm_bright = ListedColormap(['#FF0000', '#0000FF'])

plt.scatter(X_train[:, 0], X_train[:, 1], c=y_train, cmap=cm_bright, edgecolors='k')
plt.scatter(X_test[:, 0], X_test[:, 1], c=y_test, cmap=cm_bright, alpha=0.6, edgecolors='k')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from matplotlib.colors import ListedColormap

from sklearn.svm import SVC
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import SGDClassifier
from sklearn.neural_network import MLPClassifier

from sklearn.datasets import make_circles
from sklearn.datasets import make_blobs
from sklearn.datasets import make_moons

datasets = [make_moons(n_samples = 100, noise=0.2, random_state=0),
            make_circles(n_samples = 100, noise = 0.2, factor = 0.5, random_state = 0),
            make_blobs(n_samples = 100, n_features = 2, centers = 2, cluster_std = 2, random_state = 0)]

dataset_names = ['Moons', 'Circles', 'Blobs']

classifiers = [SVC(kernel = 'rbf'), 
               GradientBoostingClassifier(), 
               DecisionTreeClassifier(),
               KNeighborsClassifier(),
               RandomForestClassifier(),
               SGDClassifier(),
               MLPClassifier()]

classifier_names = ['SVC: RBF Kernel',
                    'Gradient Boosting', 
                    'Decision Tree',
                    'K Neighbors',
                    'Random Forest',
                    'Linear SGD',
                    'Neural Network']

fig, ax = plt.subplots(len(datasets),len(classifiers) + 1,figsize=(28,12))

h = 200

for ds, ds_name, i in zip(datasets, dataset_names, range(len(datasets))):
    
#     print('\t\t\t',ds_name+':\n')

    X, y = datasets[i]
    X = StandardScaler().fit(X).transform(X)
    y_named = np.array(['blue', 'red'])[y]
    X_train, X_test, y_train, y_test, y_train_named, y_test_named = \
        train_test_split(X, y, y_named, test_size = 0.4, random_state = 0)

    x_min, x_max = (X[:,0].min() - 0.5, X[:,0].max() + 0.5)
    y_min, y_max = (X[:,1].min() - 0.5, X[:,1].max() + 0.5)

    x_steps = np.linspace(x_min, x_max, h)
    y_steps = np.linspace(y_min, y_max, h)
    xx, yy = np.meshgrid(x_steps, y_steps)
    
    ax[i, 0].scatter(X_train[:,0], X_train[:,1], c = y_train_named, marker = 'o', edgecolor = 'black')
    ax[i, 0].scatter(X_test[:,0], X_test[:,1], c = y_test_named, marker = '^', edgecolor = 'black')
    ax[i, 0].set_title(ds_name)

    for clf, clf_name, j in zip(classifiers, classifier_names, range(1, len(classifiers) + 1)):

        clf.fit(X_train, y_train)

#         print(clf_name, 'Train:\n\t{:.2f}%'.format(100*clf.score(X_train, y_train)))
#         print('\t\t\t\t', clf_name, 'Test:\n\t\t\t\t\t{:.2f}%'.format(100*clf.score(X_test, y_test)))

        if hasattr(clf, 'decision_function'):
            zz = clf.decision_function(list(zip(xx.ravel(), yy.ravel()))).reshape(xx.shape)
        else:
            zz = clf.predict_proba(list(zip(xx.ravel(), yy.ravel())))[:,1].reshape(xx.shape)

        ax[i,j].contourf(xx, yy, zz, cmap = 'bwr', alpha = 0.6)

        # cmap_scatter = ListedColormap(['#00FF00', '#FF00FF'])

        ax[i, j].scatter(X_train[:,0], X_train[:,1], c = y_train_named, marker = 'o', edgecolor = 'black')
        ax[i, j].scatter(X_test[:,0], X_test[:,1], c = y_test_named, marker = '^', edgecolor = 'black')
        ax[i, j].set_title(clf_name + ': {:.1f}%'.format(100*clf.score(X_test, y_test)))
    
# plt.show()
plt.savefig('ML_algorithm_comparison.pdf', dpi = 300)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap

from sklearn.datasets import load_iris

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

ds = load_iris()
# print(ds.keys())
# print(ds.feature_names)
# print(ds.target_names)
X, y = (ds.data, ds.target)
table = pd.DataFrame(data = np.hstack((np.transpose(np.array([y])), X)), columns = np.hstack((' or '.join(ds.target_names),ds.feature_names)))
display(table)
print(ds.DESCR)
```

```python
from sklearn.ensemble import GradientBoostingClassifier

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 0)

gbrt = GradientBoostingClassifier().fit(X_train, y_train)
print('Train:\t{:.2f}%'.format(100 * gbrt.score(X_train, y_train)))
print('Test:\t{:.2f}%'.format(100 * gbrt.score(X_test, y_test)))
```

```python
gbrt = GradientBoostingClassifier(max_leaf_nodes = 5, learning_rate = 0.03, n_estimators = 100).fit(X_train, y_train)
print('Train:\t{:.2f}%'.format(100 * gbrt.score(X_train, y_train)))
print('Test:\t{:.2f}%'.format(100 * gbrt.score(X_test, y_test)))
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(max_leaf_nodes = 6, n_estimators = 200).fit(X_train, y_train)
print('Train:\t{:.2f}%'.format(100 * rf.score(X_train, y_train)))
print('Test:\t{:.2f}%'.format(100 * rf.score(X_test, y_test)))
```

```python
print(rf.predict_proba(X_test).shape)
print(rf.predict_proba(X_test)[:4,:])
print(gbrt.decision_function(X_test).shape)
print(gbrt.decision_function(X_test)[:4,:])
```

```python
from sklearn.linear_model import LogisticRegression

logreg = LogisticRegression(tol = 0.001, solver = 'lbfgs').fit(X_train, y_train)
print('Train:\t{:.2f}%'.format(100 * logreg.score(X_train, y_train)))
print('Test:\t{:.2f}%'.format(100 * logreg.score(X_test, y_test)))
```
