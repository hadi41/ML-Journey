```python
from sklearn.neighbors import KNeighborsRegressor

X2train, X2test, y2train, y2test = train_test_split(X2,y2,random_state=0)

neighbors = [n for n in range(1,10)]
accuracy = []
for n in neighbors:
    reg = KNeighborsRegressor(n)
    reg.fit(X2train, y2train)
    accuracy.append(100*reg.score(X2test, y2test))
    
fig, ax = plt.subplots(1,1,figsize=(10,6))
plt.plot(neighbors, accuracy, marker='x')
plt.xlabel('Neighbors')
plt.ylabel('Accuracy')
plt.title('Nearest Neighbor Regression')
plt.grid()
plt.show()
```

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 4))
# create 1,000 data points, evenly spaced between -3 and 3
line = np.linspace(-3, 3, 1000).reshape(-1, 1)

for n_neighbors, ax in zip([1, 3, 9], axes):
    # make predictions using 1, 3, or 9 neighbors
    reg = KNeighborsRegressor(n_neighbors=n_neighbors)
    reg.fit(X2train, y2train)
    ax.plot(line, reg.predict(line))
    ax.plot(X2train, y2train, '^', c=mglearn.cm2(0), markersize=8)
    ax.plot(X2test, y2test, 'v', c=mglearn.cm2(1), markersize=8)

    ax.set_title(
        "{} neighbor(s)\n train score: {:.2f} test score: {:.2f}".format(
            n_neighbors, reg.score(X2train, y2train),
            reg.score(X2test, y2test)))
    ax.set_xlabel("Feature")
    ax.set_ylabel("Target")
    axes[0].legend(["Model predictions", "Training data/target", "Test data/target"], loc="best")
```

```python
X_train, X_test, y_train, y_test = train_test_split(breast.data, breast.target, random_state = 0)

clf = LogisticRegression(C=0.1, max_iter = 10000)
clf.fit(X_train, y_train)
print('C=0.1 Train:\t',clf.score(X_train, y_train))
print('C=0.1 Test:\t',clf.score(X_test, y_test))
clf = LogisticRegression(C=1, max_iter = 10000)
clf.fit(X_train, y_train)
print('C=1 Train:\t',clf.score(X_train, y_train))
print('C=1 Test:\t',clf.score(X_test, y_test))
clf = LogisticRegression(C=10, max_iter = 10000)
clf.fit(X_train, y_train)
print('C=10 Train:\t',clf.score(X_train, y_train))
print('C=10 Test:\t',clf.score(X_test, y_test))
clf = LogisticRegression(C=100, max_iter = 10000)
clf.fit(X_train, y_train)
print('C=100 Train:\t',clf.score(X_train, y_train))
print('C=100 Test:\t',clf.score(X_test, y_test))
clf = LogisticRegression(C=1000, max_iter = 10000)
clf.fit(X_train, y_train)
print('C=1000 Train:\t',clf.score(X_train, y_train))
print('C=1000 Test:\t',clf.score(X_test, y_test))
clf = LogisticRegression(C=10000, max_iter = 10000)
clf.fit(X_train, y_train)
print('C=10000 Train:\t',clf.score(X_train, y_train))
print('C=10000 Test:\t',clf.score(X_test, y_test))
```
